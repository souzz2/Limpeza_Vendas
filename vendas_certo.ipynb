Substitui os fretes negativos pela mediana dos fretes positivos por cidade. Primeiro eu calculo a mediana dos fretes positivos para cada cidade. Logo após eu substituo os fretes negativos pela mediana do cidade correspondente

Aqui eu faço o cálculo da média e o desvio padrão por produto e marca. Logo depois eu mesclo as estatísticas de média e desvio padrão de volta no DataFrame original. Eu defino um intervalo de confiança para identificar outliers. Os valores que estão a mais de 3 desvios padrão da média. no final eu esxibo os produtos que são considerados outliers.

```python
media_desvio = df.groupby(['produto', 'marca'])['valor'].agg(['mean', 'std'])

df = df.merge(media_desvio, on=['produto', 'marca'], how='left')

df['outlier'] = (df['valor'] > (df['mean'] + 3 * df['std'])) | (df['valor'] < (df['mean'] - 3 * df['std']))

outliers = df[df['outlier']]

import IPython.display as display
display.display(outliers)
```

Aqui eu adiciono a coluna total calculado a coluna quantidade vezes a coluna valor, logo após eu adiciono a coluna erro total quando o total for diferente do total calculado.

```python
df['total_calculado'] = df['valor'] * df['quantidade']
df['erro_total'] = df['total'] != df['total_calculado']

print(df[df['erro_total']])
```

Aqui eu começo calculando a média e desvio padrão por produto e marca, eu também renomeio as colunas para evitar erro de KeyError. Após isso eu mesclo as esttísticas no DataFrame original. Tento identificar os outliers (valores fora de 3 desvios padrão da média) e substituo os outliers pelo valor médio do mesmo produto e marca. Depois eu removo as colunas auxiliares que não serão mais utilizadas

```python
media_desvio = df.groupby(['produto', 'marca'])['valor'].agg(['mean', 'std']).reset_index()

media_desvio.rename(columns={'mean': 'media', 'std': 'desvio'}, inplace=True)

df = df.merge(media_desvio, on=['produto', 'marca'], how='left')

df['outlier'] = (df['valor'] > (df['media'] + 3 * df['desvio'])) | (df['valor'] < (df['media'] - 3 * df['desvio']))

df.loc[df['outlier'], 'valor'] = df.loc[df['outlier'], 'media']

df = df.drop(columns=['media', 'desvio', 'outlier'])

print(df)
```

Aqui eu crio uma expressão regular para validar CEPs (5 números + hífen + 3 números). Crio uma nova coluna para marcar CEPs inválidos. Depois crio um dicionário com o CEP mais comum por cidade. Logo após substituo os CEPs inválidos pelo mais comum daquela cidade.

```python
def validar_cep(cep):
    return bool(re.fullmatch(r'\d{5}-\d{3}', cep))

df['cep_invalido'] = ~df['cep'].apply(validar_cep)

ceps_por_cidade = df[~df['cep_invalido']].groupby('cidade')['cep'].agg(lambda x: x.mode()[0])

df.loc[df['cep_invalido'], 'cep'] = df.loc[df['cep_invalido'], 'cidade'].map(ceps_por_cidade)

print(df)
```
